```python
import pandas as pd
import numpy as np
```

```python
# 读取用户表
unames=['userid','gender','age','occupation','zipcode']
users=pd.read_csv('movie-data/users.dat',delimiter="::",header=None,names=unames)
users.head(10)
```

```python
# 读取评分表
rating_name=['userid','movieid','rating','timestamp']
ratings=pd.read_csv('movie-data/ratings.dat',sep='::',header=None,names=rating_name)
print(len(ratings))
ratings.head(10)
```

```python
# 读取电影表
movie_names=['movieid','title','geners']
movies=pd.read_csv('movie-data/movies.dat',delimiter='::',header=None,names=movie_names)
print(len(movies))
print(movies.shape)
movies.head(10)
```

```python
# 合并表
data=pd.merge(pd.merge(users,ratings),movies)
data.head()
```

```python
#1.查看不同性别对电影的评分

rating_by_gender=data.pivot_table(values='rating',index='title',columns='gender',aggfunc='mean')
rating_by_gender.head(10)
```

```python
#2.不同性别差异最大的电影
rating_by_gender['diff']=np.fabs(rating_by_gender.F-rating_by_gender.M)
rating_by_gender.sort_values(by='diff',ascending=False).head(10)
```

```python
# 3.查找出热门电影  热度榜单
rating_by_title=data.groupby('title').size()
rating_by_title.sort_values(ascending=False).head(10)
```

```python
# 3.查找出热门电影  热度榜单
rating_by_title = data.groupby('title').size()
print(rating_by_title.sort_values(ascending = False).head(10))
```

```python
# 4.每部电影的评分（平均分）  高分榜单
mean_ratings = data.pivot_table(values = 'rating',index = 'title',aggfunc = 'mean')
print(mean_ratings.sort_values(by = 'rating',ascending = False).head(10))
```

```python
# 5.前十大热门电影的评分
top_10_movie = rating_by_title.sort_values(ascending = False).head(10)
mean_ratings.loc[top_10_movie.index]
```

```python
#6.前20名高分电影的热度
top_20_movie_hot = mean_ratings.sort_values(by = 'rating',ascending = False).head(20)
rating_by_title.loc[top_20_movie_hot.index]
```

```python
hot_movies = rating_by_title[rating_by_title>1000]
hot_movies_rating = mean_ratings.loc[hot_movies.index]
hot_movies_rating.sort_values(by = 'rating',ascending = False).head(20)
```
